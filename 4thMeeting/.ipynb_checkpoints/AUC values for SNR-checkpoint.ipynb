# Calculate AUC values for the file signal_to_noise_plot

```python
import numpy as np
import _pickle as pickle
import matplotlib.pyplot as plt
import os
from sklearn import metrics
```

```python
##from simulating file
#number of nodes N
N=10
#seconds for each time series simulation
#T=20
#number of time series
M=1000
```

```python
#files=['data_1.e+04.p', 'data_1.e-03.p', 'data_1.e-01.p', 'data_1.e+00.p', 'data_1.e-02.p', 'data_1.e-05.p']
files=os.listdir('/home/beate/Dokumente/00. UNI/0.Stipendium/Kolleg/network_reconstruction/4.Treffen/cubicSimulDifferentNoiseLevel/Data/.')
```

```python
#some are zeros: 1000,100000,10000 --> last 3 ones: , and 100 doesn't work for some reason ,]
files=(files[6],files[-2],files[4],files[2],files[3],files[-1],files[1],files[0],files[5],files[8],files[7])
print(files)
```

```python
#print(len(files))
expa=np.zeros((10,13))
R=np.zeros(len(files))
counter = -1
for i in range(1,len(files)):
    counter = counter + 1
    with open("/home/beate/Dokumente/00. UNI/0.Stipendium/Kolleg/network_reconstruction/4.Treffen/cubicSimulDifferentNoiseLevel/Data/%s"%files[i], "rb" ) as fd:
        data=pickle.load(fd, fix_imports=True, encoding="latin1")
        
        C = np.asarray(data)
        print(files[i])
        #print(C.shape)
        ############ for AUC
        J=np.loadtxt("/home/beate/Dokumente/00. UNI/0.Stipendium/Kolleg/network_reconstruction/4.Treffen/cubicSimulDifferentNoiseLevel/connectivitySOR.dat")
        Ad=np.copy(J)
        Ad[Ad!=0]=1

        Ad=Ad+np.eye(len(Ad))

        T=len(data)
        N,M=data[0].shape


        x=np.zeros((N,(M-1)*T)) #Zeitserie jetzt
        y=np.zeros((N,(M-1)*T)) #Zeitserie zukunft
        for t in range(T):
            x[:,t*(M-1):(t+1)*(M-1)]=data[t][:,:-1]
            x=np.nan_to_num(x, copy=True)
            y[:,t*(M-1):(t+1)*(M-1)]=data[t][:,1:]
            y=np.nan_to_num(y, copy=True)
       # print(x)    
        L=np.dot(y,np.linalg.pinv(x)) #compute pseudo inverse of x
        print('AUC step')
        AUCs=np.zeros((N,1))
        for n in range(N):
            fpr, tpr, thresholds = metrics.roc_curve(Ad[n,:],np.fabs(L[n,:]),pos_label=1)
            AUCs[n,0]=metrics.auc(fpr, tpr)

        R[i]=np.mean(AUCs)
        
        B=np.zeros((10,2000000))
        for i in range(0,10):
            a=C[:,i,:]
            a=a.reshape(1,2000000)
            B[i,:]=a
        exp=np.zeros((10,10))
        for j in range(0,10):
            a=B[j,:]
            for i in range(0,10):
                b=B[i,:]
                exp[i,j]=np.mean(np.multiply(b,a))
        for k in range(0,10):
            expa[counter,k]=exp[k,k] #diag elements
        expa[counter,11]=exp[2,7] #not diag
        expa[counter,12]=exp[6,8] #not diag
#to save aucs
np.savetxt("/home/beate/Dokumente/00. UNI/0.Stipendium/Kolleg/network_reconstruction/4.Treffen/cubicSimulDifferentNoiseLevel/AUCforsigma.dat",R,delimiter="\t",fmt="%1.4f") 
```

```python
#print(len(files))
expa=np.zeros((10,13))
R=np.zeros((19,len(files)))
counter = -1
for i in range(1,len(files)):
    counter = counter + 1
    with open("/home/beate/Dokumente/00. UNI/0.Stipendium/Kolleg/network_reconstruction/4.Treffen/cubicSimulDifferentNoiseLevel/Data/%s"%files[i], "rb" ) as fd:
        data=pickle.load(fd, fix_imports=True, encoding="latin1")
        N,M=data[0].shape
        C = np.asarray(data)
        print(files[i])
        #print(C.shape)
        ############ for AUC
        J=np.loadtxt("/home/beate/Dokumente/00. UNI/0.Stipendium/Kolleg/network_reconstruction/4.Treffen/cubicSimulDifferentNoiseLevel/connectivitySOR.dat")
        Ad=np.copy(J)
        Ad[Ad!=0]=1

        Ad=Ad+np.eye(len(Ad))
        ### for different T
        for h in range(1,20):
            T=h

            x=np.zeros((N,(M-1)*T)) #Zeitserie jetzt
            print(x.shape)
            y=np.zeros((N,(M-1)*T)) #Zeitserie zukunft
            for t in range(T):
                x[:,t*(M-1):(t+1)*(M-1)]=data[t][:,:-1]
                x=np.nan_to_num(x, copy=True)
                y[:,t*(M-1):(t+1)*(M-1)]=data[t][:,1:]
                y=np.nan_to_num(y, copy=True)
           # print(x)    
            L=np.dot(y,np.linalg.pinv(x)) #compute pseudo inverse of x
            print('AUC step')
            AUCs=np.zeros((N,1))
            for n in range(N):
                fpr, tpr, thresholds = metrics.roc_curve(Ad[n,:],np.fabs(L[n,:]),pos_label=1)
                AUCs[n,0]=metrics.auc(fpr, tpr)

            R[h-1,i]=np.mean(AUCs)
        
    
#to save aucs
np.savetxt("/home/beate/Dokumente/00. UNI/0.Stipendium/Kolleg/network_reconstruction/4.Treffen/cubicSimulDifferentNoiseLevel/AUCforsigma.dat",R,delimiter="\t",fmt="%1.4f") 
```

```python
sigma=[0.00001,0.0001,0.001,0.01,0.1,1,10,100,1000,10000,100000]
for i in range(0,19):
    plt.plot(sigma,R[i,:])
    plt.xscale('log')
plt.show()
plt.plot(sigma,R[1,:],color='green',label='short time')
plt.plot(sigma,R[10,:],color='red',label='medium time')
plt.plot(sigma,R[18,:],color='blue',label='long time')
plt.xscale('log')
plt.legend()
plt.show()
```
